By using cosine distance as a metric

```python
cosine_recommendation = pd.pivot_table(book_data,values='Book.Rating',index='Book.Title',columns='User.ID').fillna(0)
cosine_recommendation
```

```python
cosine_df = pd.DataFrame(1 - pairwise_distances(X = cosine_recommendation,metric='cosine')).round(2)
cosine_df.columns = book_data['Book.Title'].unique()
cosine_df.index   = book_data['Book.Title'].unique()
cosine_df
```
